## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## <font color=red>1. The top 10 countries with most projects</font>

```python
df_groupby_countryname = json_df.groupby(['countryname']).count()
df_groupby_countryname_sorted = df_groupby_countryname.sort_values(by=['_id'],ascending=False)
df_10 = df_groupby_countryname_sorted.head(10)
```

## <font color=red>3. Filling Missing values</font>

```python
namecode = pd.DataFrame()
namecode = json_df[['mjtheme_namecode']]
namecode.head(15)
```

```python
namecode['projects1'] = namecode['mjtheme_namecode'].str[0]
namecode['projects2'] = namecode['mjtheme_namecode'].str[1]
namecode = namecode.drop("mjtheme_namecode", axis=1)
namecode.head(5)
```

```python
namecode = namecode.fillna(0)
#namecode = namecode.replace('0', {'code': '', 'name': ''}) Does not work????
namecode.head(15)
```

## Codes and Names Table

```python
df_code_and_names = pd.DataFrame()
df_code_and_names['codes'] = [d.get('code') for d in namecode.projects1]
df_code_and_names['names'] = [d.get('name') for d in namecode.projects1]

df_code_and_names.drop_duplicates().sort_values(by=['codes'])
df_code_and_names = df_code_and_names[df_code_and_names.names != ''].sort_values(by=['codes']).drop_duplicates()
df_code_and_names = df_code_and_names.rename(columns={"codes": "Project codes", "names": "Project names"})
df_code_and_names
```

## Change mjtheme_namecode column with new filled column

```python
# build new list

mjtheme_namecode_new = list()

for i, row in namecode.iterrows():

    list_element=[row['projects1'],row['projects2']]
    mjtheme_namecode_new.append(list_element)
```

```python
se = pd.Series(mjtheme_namecode_new)
json_df['mjtheme_namecode_new'] = se.values
json_df.head(5)
```

## <font color=red>2. The top 10 major project themes (using column 'mjtheme_namecode')</font>

```python
projects = pd.DataFrame()

projects = json_df[['mjtheme_namecode_new']]

projects['projects1'] = projects['mjtheme_namecode_new'].str[0]
projects['projects2'] = projects['mjtheme_namecode_new'].str[1]
projects = projects.drop("mjtheme_namecode_new", axis=1)
projects.head(5)
```

```python
print(type(projects.projects1[0]))
print(projects.projects1[0].keys())
```

```python
projects['project1_names'] = [d.get('name') for d in projects.projects1]
projects['project2_names'] = [d.get('name') for d in projects.projects2]
#projects['project1_names']
#projects['project2_names']
projects.head()
```

```python
df1 = pd.DataFrame()
df2 = pd.DataFrame()
result = pd.DataFrame()

df1 = projects[['project1_names']]
df1 = df1.rename(columns={"project1_names": "Project_names"})
print('df1 type', type(df1))

df2 = projects[['project2_names']]
df2 = df2.rename(columns={"project2_names": "Project_names"})
print('df2 type', type(df2))

result = pd.concat([df1,df2])

print(result.shape)
print(type(result))
result = result.reset_index()
print(result.columns)
result.head()
```
